# Combining all the datasets

# Getting the cheap bench players

```python
# for each position, sort the players by cost (in ascending order)
# then, get the player with the most number of points


def get_cheapest_player(player_df):
    
    cheapest_player_names = []
    total_cost = 0
    
    for position, group in player_df.groupby('position'):
        cheapest_players =  group[(group.cost_bin == group.cost_bin.min())]
        top_cheapest_player = cheapest_players[cheapest_players.total_points_last_season == cheapest_players.total_points_last_season.max()]
        
        cheapest_player_name = top_cheapest_player.full_name.values[0]
        
        cheapest_player_names += [cheapest_player_name]
        total_cost += top_cheapest_player.cost_bin.values[0]
        
        print(position, ": ", cheapest_player_name )
        
    return cheapest_player_names, total_cost
    
```

```python
bench_players, bench_cost = get_cheapest_player(available_players_df)
```

```python
remaining_cost = total_cost - bench_cost
```

Each fantasy player starts with a limited budget of 100MM and has to buy 15 soccer players (11 for the main squad + 4 subs) in order to compete.
You need to have at least 2 Goalkeepers, 5 Defenders, 5 Midfielders, and 3 Forwards in order to complete your squad and be eligible to play.
You cannot have more than 3 players from the same soccer team.


* Check if a player is injured, suspended or unavailable to play and if so, do not pick that player for our Fantasy team even if they have a high ROI.
* Pick the top three star players with the most cumulative league points in the league first. (We will test the outcome of this condition with different numbers of star-players and pick the version that generates the biggest return on investment, while still allowing enough left-over budget to fill our squad with a lot of the Top 50 ROI players.)
* Every time we pick a player and add them to our team, we subtract their cost from our 100MM budget and we add their position and team-name to a list, to make sure that we stop buying players for the positions and the teams that hit their constraint limit.
* Once the optimal number of expensive superstar players are picked, the Algorithm starts going through the list of players with the highest ROI and tries to get us as many of the top names as it can, until we get close to depleting our budget and filling all of the team positions.
* Algorithm prints a list of the players it picked at the end and gives us the remaining budget and the total fantasy points of the team.

## Getting players with top performance from available players

```python
# since bench has 4 players remaining is 11 players and the remaining cost is obtained. now we set the constraints and get the remaining team.

most_points_players = available_players_df[['full_name','team_name','position','total_points_last_season','cost_bin']]
most_points_players = most_points_players.sort_values(by=['total_points_last_season'],ascending=False)
    
```

```python
most_points_players.head()
```

# Calculating the player ROI

```python
# next we create players with highest ROI 

roi_players = available_players_df[['full_name','team_name','position','total_points_last_season','cost_bin']]
```

```python
roi_players['roi'] = roi_players.apply(lambda row: row.total_points_last_season / row.cost_bin, axis=1)
```

```python
roi_players = roi_players.sort_values(by= 'roi', ascending = False)
roi_players.head()
```

```python
roi_players = roi_players.dropna()
```

```python
#roi_players.to_csv('roi.csv')
```

```python
def money_team():
    roi_team = []
    total_points = 0
    budget = remaining_cost
    top_performer_limit = 2
    position_dict = {"Keeper": 1, "Defender": 4, "Midfielder": 4, "Forward": 2}
    team_dict = {'Arsenal':3, 'Aston Villa':3,'Brentford':3, 'Brighton':3,'Burnley':3,'Chelsea':3,
                 'Crystal Palace':3,'Everton':3,'Leicester':3,'Leeds':3,'Liverpool':3,'Man City':3,
                 'Man Utd':3,'Newcastle':3,'Norwich':3,'Southampton':3,'Spurs':3,'Watford':3,'West Ham':3,'Wolves':3}
    
    # top 2 performers
    
    for idx, row in most_points_players.iterrows():
        if budget >= row.cost_bin and len(roi_team)<top_performer_limit and position_dict[row.position] != 0 and team_dict[row.team_name] != 0:
            roi_team.append(row.full_name)
            budget -= row.cost_bin #Deducting cost from budget
            position_dict[row.position] -= 1 # Deducting position from position dictionary
            team_dict[row.team_name] -= 1 # Deducting player from team dictionary
            total_points += row.total_points_last_season # adding to point score
            print("Player choosen from 'top players' " + str(row.full_name))
    
        # Choosing the remaining team from ROI data
        
        else:
            for idx, row in roi_players.iterrows():
                if row.full_name not in roi_team and budget >= row.cost_bin and position_dict[row.position] != 0 and team_dict[row.team_name] != 0:
                    roi_team.append(row.full_name)
                    budget -= row.cost_bin
                    position_dict[row.position] -= 1 # Deducting position from position dictionary
                    team_dict[row.team_name] -= 1 # Deducting player from team dictionary
                    total_points += row.total_points_last_season # adding to point score
    print("\nTeam chosen: " + str(roi_team))
    print("Remaining budget: " + str((budget)) + "M.")
    print("Total points from choosen team: " + str(total_points) + ".")
    print(position_dict)
        
    
```
